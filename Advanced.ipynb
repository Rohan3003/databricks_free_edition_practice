```python
large_data = [(i, f"name_{i}") for i in range(1, 1000001)]

large_df = spark.createDataFrame(large_data, ["id", "name"])

small_data = [
    (1, "HR"),
    (2, "IT"),
    (3, "Finance"),
    (4, "Admin")
]

small_df = spark.createDataFrame(small_data, ["id", "department"])
```

```python
from pyspark.sql.functions import broadcast
joined_df = large_df.join(broadcast(small_df), on = "id", how = "inner")
joined_df.count()
```

```python
joined_df.show()
```

# Execution Plan

```python
# Non optimized join
large_df.join(small_df, on = "id", how = "inner").explain()
```

```python
# Optimized join - broadcast
from pyspark.sql.functions import broadcast
large_df.join(broadcast(small_df), on = "id", how = "inner").explain()
```
