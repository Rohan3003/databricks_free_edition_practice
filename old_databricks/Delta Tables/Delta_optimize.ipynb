```python
from pyspark.sql.functions import col


schema_emp = ['emp_id', 'emp_name', 'dept_name', 'dept_id']
schema_dept = ['deptID', 'dept_name']

data_emp = [(1, 'jack', 'HR', 10), (2,'Vick', 'IT', 11), (3, 'George', 'Admin', None)]
data_dept = [(11, 'IT'), (10, 'HR')]

df_emp = spark.createDataFrame(data = data_emp, schema = schema_emp)
df_dept =spark.createDataFrame(data = data_dept, schema = schema_dept)

df_emp.show()
df_dept.show()
```

```python
df_emp.join(df_dept).show()
```

```python
df_emp.join(df_dept).explain()
```

```python
df_emp.join(df_dept, df_emp.dept_id == df_dept.dept_id).show()
```

```python
df_emp.join(df_dept, df_emp.dept_id == df_dept.dept_id).explain()
```

```python
# to check type of join preferred -> if True - sortMergeJoin or if False
spark.conf.get("spark.sql.join.preferSortMergeJoin")
```
