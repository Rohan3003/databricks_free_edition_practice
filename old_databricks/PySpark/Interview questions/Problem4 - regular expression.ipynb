```python
# Imports
from pyspark.sql.functions import lower, col, regexp_replace, length
```

```python
data = [(1,"I love to play cricket"),(2, "I am into motorbiking"),(3,"What  do you like")]
schema = ["id", "message"]
df = spark.createDataFrame(data = data, schema = schema)
df.show(truncate = False# f)
```

```python
# find the number of consonants 
df_consonants = df.withColumn("consonants", lower(col("message")))
df_consonants.show(truncate = False)
```

```python
# regexp_replace(column, "expression", "with what")
df_consonants = df_consonants.withColumn("message_without_space",\
    regexp_replace(col("consonants"),"a|e|i|o|u|\s",""))
df_consonants.show(truncate = False)
```

```python
# length function is used to count the number of character in a columns
df_answer = df_consonants.withColumn("count",length(col("message_without_space")))
df_answer.show(truncate = False)
```

```python
df_answer = df_answer.withColumnRenamed("count","consonant counts")
df_answer.show(truncate = False)
```
