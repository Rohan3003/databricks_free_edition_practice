Problem 1) get  count the number of activities performed by each user per day in a time-series dataset.
columns: useri, activityid, date

```python
from pyspark.sql.types import StructType, StructField, IntegerType, DateType
import datetime
import random
schema = StructType([
    StructField('userid', IntegerType(), True),
    StructField('activityid', IntegerType(), True),
    StructField('date', DateType(), True)
])
# Create sample data - 20 records
base_date = datetime.date(2025, 5, 18)
user_ids = [101, 102, 103, 104]  # Example integer user IDs

data = []
for i in range(10):
    userid = random.choice(user_ids)+1
    activityid = (i % 5) + 1                # activityid as integer: 1 to 5
    date = base_date + datetime.timedelta(days=i // 4)
    data.append((userid, activityid, date))

# Create DataFrame
df = spark.createDataFrame(data, schema)
df.show()
```

```python
df.limit(5).show()
```

```python
# to get number of activities per user id
from pyspark.sql.functions import count
df.groupBy("userid").agg(count("*").alias("activity_count")).show()
```

```python
# to get number of activities per user id per day
from pyspark.sql.functions import count
df.groupBy("userid","date").agg(count("*").alias("activity_count")).orderBy("userid").show()
```
