```python
#from pyspark.sql.functions import
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, col


w = Window.partitionBy("orderID").orderBy("orderID")
df_dup = df.withColumn("rn", row_number().over(w))
#df_dup.show()
df_dedup = df_dup.select("Region","Country","ItemType","SalesChannel","OrderPrice","OrderDate","OrderID")\
    .filter(col("rn") < 2)
df_dedup.show()
```
