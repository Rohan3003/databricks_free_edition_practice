```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

# Specify the folder path where the CSV files are located
folder_path = "dbfs:/FileStore/tables/"

# Specify the file names you want to read
files_to_read = ["Employee1.csv", "Employee2.csv"]

# Read the specific CSV files
df = spark.read.csv(folder_path, header=True, inferSchema=True, ignoreLeadingWhiteSpace=True, ignoreTrailingWhiteSpace=True, **{"pathGlobFilter": "|".join(files_to_read)})
```

##### to define a schema of a csv file


```python
from pyspark.sql.types import * 
help(StructType)
```

```python
from pyspark.sql.types import StructType, StringType, IntegerType

schema = StructType().add(field='id',data_type=IntegerType())\
                    .add(field='name', data_type=StringType())\
                    .add(field='gender', data_type=StringType())\
                    .add(field='salary', data_type=IntegerType())

df = spark.read.csv('dbfs:/FileStore/tables/Employee1.csv', schema = schema, header =True)
display(df)
df.printSchema()
```

```python
path = '/databricks-datasets/COVID/USAFacts/*.csv'
df_csv_schema = spark.read.format("csv") \
                   .option("header", "True") \
                   .option("inferSchema", "True") \
                   .load(path)
```
