#### To select rows of a columns into a list

1. Method1 - using df.rdd.map()
2. Method2 - using i.Row() method while looping

```python
data = [("James","Smith","USA","CA"),\
        ("Michael","Rose","USA","NY"), \
        ("Robert","Williams","USA","CA"),\
        ("Maria","Jones","USA","FL")]

columns=["firstname","lastname","country","state"]
df=spark.createDataFrame(data=data,schema=columns)
df.show()
df.collect()
```

```python
firstname_list = df.rdd.map(lambda x : x[0]).collect()
lastname_list = df.rdd.map(lambda x : x [1]).collect()
country_list = df.rdd.map(lambda x : x [2]).collect()
state_list = df.rdd.map(lambda x : x [3]).collect()

print(firstname_list)
print(lastname_list)
print(country_list)
print(state_list)
```

```python
df.show()
l1 = []
for i in df.select('state').collect():
    l1.append(i.state)
print(l1)

l2 = []
l3 = []
l4 = []
for i in df.select('*').collect():
    l2.append(i.firstname)
    l3.append(i.lastname)
    l4.append(i.country)

print(l2)
print(l3)
print(l4)
```
