### Creating dataframes with defined schema

```python
# Define the schema for the dataframe
schema = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("gender", StringType(), True)
])

# Define a list of Python dictionaries with random data
data = [
    {"name": "Alice", "age": 25, "gender": "F"},
    {"name": "Bob", "age": 30, "gender": "M"},
    {"name": "Charlie", "age": 20, "gender": "M"},
    {"name": "Dave", "age": 35, "gender": "M"},
    {"name": "Eve", "age": 28, "gender": "F"}
]

# Create the dataframe from the list of dictionaries
df1 = spark.createDataFrame(data, schema)

# Add a random column to the dataframe
df1 = df1.withColumn("random", rand())
df1.show()
df1.printSchema()
```

### using Row and schema 

```python
myManualSchema = StructType([
                StructField('some', StringType(), nullable=True),\
                StructField('col', StringType(), nullable=True),\
                StructField('names', LongType(), nullable=True)
                            ])
myRow = Row('Hello', None, 2)
myDf = spark.createDataFrame([myRow], myManualSchema)
myDf.show()
```

### dataframe with all possible datatypes

```python
schema = T.StructType([
    T.StructField("col_string", T.StringType()),
    T.StructField("col_integer", T.IntegerType()),
    T.StructField("col_long", T.LongType()),
    T.StructField("col_float", T.FloatType()),
    T.StructField("col_double", T.DoubleType()),
    T.StructField("col_boolean", T.BooleanType()),
    T.StructField("col_date", T.DateType()),
    T.StructField("col_timestamp", T.TimestampType()),
    T.StructField("col_array", T.ArrayType(T.StringType())),
    T.StructField("col_map", T.MapType(T.StringType(), T.IntegerType())),
    T.StructField("col_struct", T.StructType([
        T.StructField("sub_col1", T.StringType()),
        T.StructField("sub_col2", T.IntegerType())
    ]))
])

data = [
    ("string_value", 123, 456789, 1.23, 4.56, True, None, None,
     None, {"map_key": 1}, ("sub_value1", 789))
]

df_all_dtypes = spark.createDataFrame(data, schema)
df_all_dtypes.display()
df_all_dtypes.dtypes
```
