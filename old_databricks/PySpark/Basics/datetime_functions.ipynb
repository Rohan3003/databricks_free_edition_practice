### Date and time in pyspark
1. to_date()
2. date_format()
3. current_timestamo()
4. strftime()

```python
from pyspark.sql.functions import current_date, current_timestamp, to_utc_timestamp, col, datediff
```

```python
data=[["1"]]
df_date = spark.createDataFrame(data,["id"])

df_date.withColumn("current_date", current_date())\
       .withColumn("current_timestamp", current_timestamp()) \
       .withColumn("current_timestamp_india", to_utc_timestamp("current_timestamp","IST")) \
       .withColumn("current_timestamp_australia", to_utc_timestamp("current_timestamp","Australia/Sydney")) \
       .display()
```

```python
# The timestamp appears 1 hour ahead of UTC since daylight saving is enabled in the summer
df_date.select(to_utc_timestamp("current_timestamp","IST").alias("utc_time")).display()
```

### datediff()

```python
data = [("1","2023-01-01"),("2","2023-02-24"),("3","2023-03-24")]
df = spark.createDataFrame(data=data,schema=["id","date"])
df.display()
```

```python
df.select(col("date"), current_date().alias("current_date"), datediff(current_date(), col("date")).alias("date_diff")).display()
```
