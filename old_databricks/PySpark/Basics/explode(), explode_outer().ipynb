The following methods are used when we work with ArrayType columns in dataframes :
  1. explode()
  2. explore_outer()
  3. posexplode()
  4. posexplode_outer()
  5. split()
  6. array()
  7. array_contains()
  8. array_zip()

```python
# imports statements
from pyspark.sql.functions import col, explode, explode_outer, posexplode, posexplode_outer, split, array, array_contains
```

```python
arrayData = [
        ('James',['Java','Scala'],{'hair':'black','eye':'brown'},['Bangalore','San Jose']),
        ('Michael',['Spark','Java',None],{'hair':'brown','eye':None},[None]),
        ('Robert',['CSharp',''],{'hair':'red','eye':''},[None,'San Francisco']),
        ('George',None,None,None),
        ('Jefferson',['1','2'],{},[])
            ]

schema = ['name', 'knownLanguages', 'properties', 'location']
df = spark.createDataFrame(data = arrayData, schema = schema)
df.printSchema()
df.show(truncate=False)
```

```python
df.display()
```

### explode()

```python
# explode array column - create new row for same data with new col 
df_array_column1 = df.select(col("name"),explode(col("knownLanguages")).alias("known_languages"))
df_array_column1.show()
```

Explode map columns

```python
# explode map column - breaks input column to key value pairs
df_map_column = df.select(col("name"),explode(col("properties")))
df_map_column.show(truncate = False)
```

### array_zip()

```python
# same datatype -> array can be exploded together using array_zip()
from pyspark.sql.functions import arrays_zip, explode

df.select("name", "properties",
          explode(arrays_zip("knownLanguages", "location")).alias("zipped")).select("name", "properties", "zipped.knownLanguages", "zipped.location").show()
```

```python
df.select("name", "properties",
          explode_outer(arrays_zip("knownLanguages", "location")).alias("zipped")).select("name", "properties", "zipped.knownLanguages", "zipped.location").show()
```
