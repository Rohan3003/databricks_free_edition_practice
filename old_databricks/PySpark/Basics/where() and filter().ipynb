```python
import pyspark.sql.functions as F
from pyspark.sql.functions import col
```

```python
data = [(1,"Bangalore",140)
        ,(2, "Gurgaon", 12), 
        (3,"Chennai",120), 
        (4,"Pune",73),
        (1,"Tumkur",4), 
        (2,"Noida", 9)]
schema = ["state_code","city","population (lakhs)"]
df = spark.createDataFrame(data=data, schema=schema)
df.show()
```

```python
df_karnataka = df.filter(F.col("state_code") == 1)
df_karnataka.show()
```

```python
# to exclude cities with population less than and equal to 12 lakhs
df.filter(F.col('population (lakhs)') <= 12).show()
filter_df = df.filter(~(F.col('population (lakhs)') <= 12) & ~(F.col('city').startswith('B')))
```

```python
filter_df.show()
```

#### 1. filter() based on list

```python
age_list = [25,20]
df_customer.filter(df_customer.age.isin(age_list)).show()
```

#### 2. filter() with single conditions

```python
df_customer.filter(col("name") == "Bob Victor Dsouza").show()
```

#### 3. filter() with mutiple conditions

```python
df_customer.filter((col("age")>28) & (col("gender") == 'M')).show()
```

#### 4. startswith(), endswith() & contains()

```python
df_customer.filter(col("name").startswith("B")).show()
```

```python
df_customer.filter(col("name").endswith("e")).show()
```

```python
df_customer.filter(col("name").contains("Bob")).show()
```
