1. coalesce() function returns first columns from the provided columns with non-null values
2. It has to be imported from pyspark.sql.functions
3. The function takes 2 parameter
      coalesce(col, default_value)

```python
from pyspark.sql.functions import coalesce, lit, array
df = spark.createDataFrame([(None,None), (9,None), [4,2], (None, 29)], ("col1", "col2"))
df.show()
```

```python
df.withColumn('coalesce_value',coalesce(df["col1"], df["col2"])).show()
```

```python
df.withColumn('coalesce_value',coalesce(df["col1"], lit(1))).show()
```

## coalesce()

```python
df_coalesce = spark.read.format("csv")\
                        .option("header","true")\
                        .option("inferSchema","true")\
                        .load("dbfs:/FileStore/PySpark concepts/simple_zipcodes.csv")
df_coalesce.printSchema()
```

```python
df_coalesce.show()
```

```python
# setting the number of partition to 10 
print("Number of partitions before repartition : ", df_coalesce.rdd.getNumPartitions())
df_coalesce = df_coalesce.repartition(10)
print("Number of partitions after repartition (10) :", df_coalesce.rdd.getNumPartitions())
```

```python
# number of partiton after the coalesce 
df_coalesced = df_coalesce.coalesce(2)
df_coalesced.rdd.getNumPartitions()
```

```python
df_coalesced.write.format("csv").mode("overwrite").save("dbfs:/FileStore/PySpark concepts/simple_zipcodes/coalesced_data")
```

## repartition()

```python
df_repartition = spark.read.format("csv")\
                           .option("header","true")\
                           .option("inferSchema","true")\
                           .load("dbfs:/FileStore/PySpark concepts/simple_zipcodes.csv")
df_repartition.printSchema()
```

```python
print("Initial Number of partitions are",df_repartition.rdd.getNumPartitions())
new_df = df_repartition.repartition(3)
print("The number of partitions are now :",new_df.rdd.getNumPartitions())
```

```python
new_df.display()
```

```python
new_df.groupBy('State').count().alias('No of records in each State').show()
```
