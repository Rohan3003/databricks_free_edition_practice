### groupBy()
This statement is used to collect() identical data into groups on a dataframe and perform the following aggregation
1. count()
2. sum()
3. avg()
4. min()
5. max()

```python
data = [("James","Sales","NY",90000,34,10000),
    ("Michael","Sales","NY",86000,56,20000),
    ("Robert","Sales","CA",81000,30,23000),
    ("Maria","Finance","CA",90000,24,23000),
    ("Raman","Finance","CA",99000,40,24000),
    ("Scott","Finance","NY",83000,36,19000),
    ("Jen","Finance","NY",79000,53,15000),
    ("Jeff","Marketing","CA",80000,25,18000),
    ("Kumar","Marketing","NY",91000,50,21000)
  ]

schema = ["employee_name", "department", "state", "salary", "age", "bonus"]
df = spark.createDataFrame(data=data, schema = schema)
df.printSchema()
df.show(truncate=False)
```

```python
# groupBy operation returns GroupData type object
type(df.groupBy("depatment"))
```

#### count() 
count does not take any argument

```python
df_groupBy = df.groupBy("department").count().withColumnRenamed('count','No of employees in each department')
df_groupBy.show()
```
