##### StructType() and StructField()
These functions are used to programmatically specify schema to a dataframe in pyspark.
They are used in columns types which are 
  * nested type 
  * map columns - {'key1': 'value1', 'key2' : 'value2'}
  * array type - ['value1', 'value2']
  
**StructType** 
1. StructType is used to define structure of dataframe
2. StructType is a class which contains a collection of  StructField.

**StructField**
1. StructField is used to define the metadata of dataframe columns 
2. StructField is used to define column name, column datatype, boolean to specify if column can have nullable values or not. if True = nullable, if False = not nullable


```python
arrayData = [
        ('James',['Java','Scala'],{'hair':'black','eye':'brown'}),
        ('Michael',['Spark','Java',None],{'hair':'brown','eye':None}),
        ('Robert',['CSharp',''],{'hair':'red','eye':''}),
        ('Washington',None,None),
        ('Jefferson',['1','2'],{})
            ]

schema = ['name', 'knownLanguages', 'properties']
df = spark.createDataFrame(data = arrayData, schema = schema)
df.printSchema()
df.show()
```

```python
from pyspark.sql.functions import explode
df2 = df.select('knownLanguages', explode(df.knownLanguages))
```
