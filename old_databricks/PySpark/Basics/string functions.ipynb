```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

data = [  (1, 'John', 'Doe', 'john.doe@example.com', 'New York'),  (2, 'Jane', 'Doe', 'jane.doe@example.com', 'Los Angeles'),  (3, 'James', 'Smith', 'james.smith@example.com', 'Chicago'),  (4, 'mary', 'Johnson', 'mary.johnson@example.com', 'Houston')]

schema = StructType([
  StructField('id', IntegerType(), True),
  StructField('firstname', StringType(), True),
  StructField('lastname', StringType(), True),
  StructField('email', StringType(), True),
  StructField('address', StringType(), True)
])

df = spark.createDataFrame(data, schema)
```

```python
df.show(truncate=False)
```

```python
# to concat two columns
# ex- John | Doe --> John Doe
from pyspark.sql.functions import concat, col, lit
df.select(concat(col('firstname'),lit(' '),col('lastname'))).alias('fullname').show(truncate=False)
```

```python
# to find the domain name from email column
# ex - john.doe@example.com --> example.com
from pyspark.sql.functions import substring
df.withColumn('domain_name').su
```
