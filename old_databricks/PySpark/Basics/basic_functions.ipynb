### alias()
alias() method is used to create an alias in PySpark for a column, DataFrame and SQL table

Use - 

1. Column aliasing - Aliasing a PySpark dataframe column changes the name of the column without changing the datatype and the data.
2. DataFrame aliasing - Aliasing a DataFrame gives a different name to the DataFrame without changing the structure, data and column names of the DataFrame.
3. SQL Table aliasing - Aliasing a SQL Table gives a different name to the SQL table without changing the structure, data and column names of the table.


```python
df_emp.alias('employee')
```

```python
df_emp.select("*",F.col("bonus").alias("salary_bonus")).show()
```

### Pivot()

It is used to transpose the data from one column into mutiple columns

```python
data = [("Banana",1000,"USA"), ("Carrots",1500,"USA"), ("Beans",1600,"USA"), \
      ("Orange",2000,"USA"),("Orange",2000,"USA"),("Banana",400,"China"), \
      ("Carrots",1200,"China"),("Beans",1500,"China"),("Orange",4000,"China"), \
      ("Banana",2000,"Canada"),("Carrots",2000,"Canada"),("Beans",2000,"Mexico")]

columns= ["Product","Amount","Country"]
df_item = spark.createDataFrame(data = data, schema = columns)
df_item.printSchema()
df_item.show(truncate=False)
```

```python
df_item.groupBy("Product").sum("Amount").show()
```

```python
df_item.groupBy("Product").pivot("Country").sum("Amount").show()
```

```python
data = [('India', 100, 'Banana'),('India', 200, 'Orange'),('China', 300, 'Banana'),('China', 400, 'Orange')]
schema = ['Country', 'Amount', 'Fruits']
df_country = spark.createDataFrame(data = data, schema = schema)
df_country.show()
```

### withColumn() and cast()
**withColumn()**
1. withColumn() is a transformation function of DataFrame which is used to manipulate a column
2. convert the datatype of an existing column 
3. create a new column of dataframe


**cast()**
1. It is used to provide datatype to an columns while using withColumn 

```python
df_emp.printSchema()
```

```python
df_salary = df_emp.withColumn("salary per month", col("salary")/100)
df_salary.printSchema()
df_salary.show()
```

```python
# without cast
df_salary_total = df_emp.withColumn("total income", (col("salary")+ col("bonus")))
df_salary_total.show()
df_salary_total.printSchema()
```

```python
# with cast 
df_salary_total = df_emp.withColumn("total income", (col("salary")+ col("bonus")).cast("Integer"))
df_salary_total.show()
df_salary_total.printSchema()
```

### collect()
1. It is an action in pyspark to retrieve all the elements of dataset from all workder nodes to driver node
2. Ideally, we should use collect() on small datasets after applying filter() or where() operation
3. Retrieving large datasets results in **OutOfMemory** issue


collect() command retrieves all the elements in a DataFrame as **an Array of RowType** to the driver node

Once you have performed collect(), you can use python loop to iterate over them

```python
collect = df_customer.collect()
print(collect)
```

```python
# to put all the names in a list 
names = []
for i in collect:
    names.append(i.name)
    
names
```

```python
# to put all the ages in a list
age = []
for j in collect:
    age.append(j.age)
age
```

```python
name_age = zip(names, age)
type(name_age)
```

```python
l1, l2 = list(zip(*name_age))
print(list(l1))
print(l2)
```
