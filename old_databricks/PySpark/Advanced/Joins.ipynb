```python
# join on mutiple columns with example
data1 = [(1, "A"), (2, "B"), (3, "C")]
columns1 = ["A1", "A2"]
df1 = spark.createDataFrame(data1, columns1)
df1.show()

data2 = [(1, "F"), (2, "B")]
columns2 = ["B1", "B2"]
df2 = spark.createDataFrame(data2, columns2)
df2.show()

# Perform a join on multiple columns
join_condition = [df1["A1"] == df2["B1"], df1["A2"] == df2["B2"]]
df = df1.join(df2, on=join_condition, how='inner')

# Show the resulting DataFrame
df.show()
```

```python
# what happens if you join dataframes without matching values of the key column
data1 = [(1, "A"), (2, "B"), (3, "C")]
columns1 = ["A1", "A2"]
df1 = spark.createDataFrame(data1, columns1)
df1.show()

data2 = [(7, "F"), (6, "B")]
columns2 = ["B1", "B2"]
df2 = spark.createDataFrame(data2, columns2)
df2.show()

# Perform a join on multiple columns
join_condition = [F.col("A1") == F.col("B1")]
df = df1.join(df2, on = join_condition, how='inner')

# Show the resulting DataFrame
df.show()
```

```python
# what happens if you join dataframes without matching values of the key column because of datatype
data1 = [(1, "A"), (2, "B"), (3, "C")]
columns1 = ["A1", "A2"]
df1 = spark.createDataFrame(data1, columns1)
df1.show()
df1.printSchema()

data2 = [("1", "F"), ("2", "B")]
columns2 = ["B1", "B2"]
df2 = spark.createDataFrame(data2, columns2)
df2.show()
df2.printSchema()

# Perform a join on multiple columns
join_condition = [F.col("A1") == F.col("B1")]
df = df1.join(df2, on = join_condition, how='inner')

# Show the resulting DataFrame
df.show()
```
