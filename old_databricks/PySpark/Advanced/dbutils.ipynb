### dbutils.widgets


* text - input a variable in text box
* drop down - select a value from a list of dropdown
* combobox - combination of text and dropdown. Select a value from a provided list of input ones in text box
* multiselect - select one or more values from a list of provided values

#### dbutils.widgets.text()

```python
# Text
dbutils.widgets.removeAll()
dbutils.widgets.text("Text", "Hello world")
```

#### dbutils.widgets.dropdown()

```python
help(dbutils.widgets.dropdown)
```

```python
# Dropdowm menu
dbutils.widgets.dropdown("Dropdown", "1", [str(x) for x in range(1,10)], "Dropdown label")
```

```python
dbutils.widgets.dropdown("Dropdown2"," ",["a","b"," "],label="Dropdown label2")
```

```python
,#
, Dropdown widget with label name : 1. List Paths -> contains values from list_path and empty string [""] also sorted in alphabetical order
list_paths = ['z', 'c', 'a']
dbutils.widgets.dropdown('list_paths',"",[""] + sorted(list_paths), "1.List Paths")
var_path = dbutils.widgets.get("list_paths")
print(var_path)
```

#### dbutils.widgets.combobox()

```python
# Combobox 
dbutils.widgets.combobox("Combobox", "A", ["A", "B", "C"])
```

```python
# Multiselect 
dbutils.widgets.multiselect("Multiselect", "Yes", ["Yes", "No", "Maybe"])
```

#### dbutils.fs.mv()
dbutils.fs.mv(location1, location2)

1. Moves a file or directory, from one path to another
2. It has two steps <br>
          1. copy files/directory from location1 to location2 <br>
          2. delete files/directory from location1

```python
display(dbutils.fs.ls('/user/delta_table'))
```

```python
dbutils.fs.mv('/user/delta_table','/user/', True)
```

```python
display(dbutils.fs.ls('/user/delta_table'))
```

```python
display(dbutils.fs.ls('/user/'))
```

```python
file_format = 'parquet'
file = [file_parquet for file_parquet in dbutils.fs.ls('/user') if file_parquet.name.endswith('.'+ file_format)]
```

```python
for files in file:
    print(str(files.name))
```

```python
for files in file:
    dbutils.fs.mv(files.path,'/user/delta_table', True)
```
