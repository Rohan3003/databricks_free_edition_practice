```python
from pyspark.sql.functions import broadcast
```

```python
# Creating fact table 
transaction = [
    (100, 'Cosmetic', 150),
    (200,'Apparel',250),
    (300,'Shirt',400),
    (400,'Trouser',500),
    (500,'Socks',20),
    (100,'Belt',70),
    (200,'Cosmetic',250),
    (300,'Shoe',400),
    (400,'Socks',25),
    (500,'Shorts',100)
]
df_transaction = spark.createDataFrame(data = transaction, schema = ['store_id', 'item', 'amount'])
df_transaction.show()
```

```python
# Creating dimension table
store = [
    (100, 'store_london'),
    (200, 'store_paris'),
    (300, 'store_frankfurt'),
    (400, 'store_stockholm'),
    (500, 'store_oslo')
]

df_store = spark.createDataFrame(data = store, schema = ['store_id', 'store_name'])
df_store.show()
```

```python
# inner join without using broadcast variable
df_join = df_transaction.join(df_store, df_transaction['store_id'] == df_store['store_id'])
df_join.show()
df_join.explain()
```

```python
df_join_broadcast = df_transaction.join(df_store, df_transaction['store_id'] == df_store['store_id'])
df_join_broadcast.show()
df_join_broadcast.explain()
```
