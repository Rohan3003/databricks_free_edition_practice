```python
data = [1, 2, 2, 3, 4, 5, 5, 5, 6]
print(type(data))

# to create a rdd use parallelize function
rdd = sc.parallelize(data)
print(type(rdd))
```

```python
# map() transformation
rdd_map = rdd.map(lambda x : x*2)
print(rdd_map)

# since transformation are not written to a driver node, we need to use take() action 
rdd_map.take(10)
```

```python
# filter()
rdd_filter = rdd.filter(lambda z : z >= 5)
rdd_filter.take(10)
```

```python
# distinct() - to get the unique values of the RDD only
rdd_distinct = rdd.distinct()
rdd_distinct.take(10)
```

```python
# union() - this creates a new RDD which contains all elements from both rdd and rdd2
rdd2 = sc.parallelize([5,5,6,7,8,9,9])
rdd_union = rdd.union(rdd2)
rdd_union.take(20)
```

```python
# intersection - this creates a new RDD with distinct intersecting values from rdd and rdd2 
rdd_intersec = rdd.intersection(rdd2)
rdd_intersec.take(10)
```

```python
# subtract() - creates a new rdd which has elements from first rdd but not in second RDD
print(rdd.take(10))
print(rdd2.take(10))
rdd_sub = rdd.subtract(rdd2)
rdd_sub.take(10)
```

```python
#sample()
print(rdd.take(10))
rdd_sample = rdd.sample(False, 0.5)
rdd_sample.collect()
```
