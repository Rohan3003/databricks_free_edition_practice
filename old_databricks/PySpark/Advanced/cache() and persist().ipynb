```python
########################################################################### df_emp #######################################################################################
# Creating dataframe without defining any schema

data1 = [("James","Sales","NY",90000,34,10000), \
    ("Michael","Sales","NY",86000,56,20000), \
    ("Robert","Sales","CA",81000,30,23000), \
    ("Maria","Finance","CA",90000,24,23000), \
    ("Raman","Finance","CA",99000,40,24000), \
    ("Scott","Finance","NY",83000,36,19000), \
    ("Jen","Finance","NY",79000,53,15000), \
    ("Jeff","Marketing","CA",80000,25,18000), \
    ("Kumar","Marketing","NY",91000,50,21000) \
  ]
columns= ["employee_name","department","state","salary","age","bonus"]
df_emp = spark.createDataFrame(data = data1, schema = columns)
df_emp.printSchema()


###################################################################### df_customer ###############################################################################################
# Creating dataframe with defined schema 
schema = T.StructType([T.StructField("name", T.StringType(), True), \
                     T.StructField("age", T.IntegerType(), True), 
                     T.StructField("gender", T.StringType(), True),\
                     T.StructField("phone number", T.IntegerType(), True)
])

# Define a list of Python dictionaries with random data
data = [{"name": "Alice Sansumg Israel", "age": 25, "gender": "F", "phone number" : 983529868}, \
        {"name": "Bob Victor Dsouza", "age": 30, "gender": "M", "phone number" : 919951557}, \
        {"name": "Charlie", "age": 20, "gender": "M", "phone number" : 13121321},\
        {"name": "Dave", "age": 35, "gender": "M", "phone number" : 2137921398},\
        {"name": "Eve", "age": 28, "gender": "F", "phone number" : 3121232},\
        {"name": "Bob Singh Dsouza", "age": 31, "gender": "F", "phone number" : 12121}]

# Create the dataframe from the list of dictionaries
df_customer = spark.createDataFrame(data, schema)
df_customer.printSchema()

df_emp.show()
df_customer.show()
```

```python
# cache() dataframe
df_cache = df_emp.cache()
```

```python
# count() took [1.3 seconds] which is approx 3 times count() with caching
df_emp.count()
```

```python
# count() with cahche took [0.5 seconds], certainely cache operation makes this faster
df_cache.count()
```

### persist()
1. persist() is also another **spark optimization** technique to store data either in **memory** or **disc** across various **worker nodes**.
2. Storage level for persist()
    * MEMORY_ONLY - same as cache()
    * MEMORY_ONLY_SER
    * MEMORY_AND_DISK
    * MEMORY_AND_DISK_SER
    * DISK_ONLY

```python
df_customer.explain()
```

```python
# perist() dataframe
df_customer_persist1 = df_customer.persist()
```

```python
df_customer.count()
```

```python
df_customer_persist1.count()
```

```python
df_customer_persist1.explain()
```

```python
from pyspark.storagelevel import StorageLevel
df_customer_persist2 = df_customer.persist(StorageLevel.MEMORY_AND_DISK)
df_customer_persist2.count()
```

```python
df_customer_persist2.explain()
```

```python
df = df_customer_persist2.persist(StorageLevel.MEMORY_ONLY)
df.explain()
```
