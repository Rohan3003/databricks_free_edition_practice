```python
from pyspark.sql.functions import col, lit, row_number, dense_rank, rank
from pyspark.sql.window import Window

simpleData = (("James", "Sales", 3000), \
    ("Michael", "Sales", 4600),  \
    ("Robert", "Sales", 4100),   \
    ("Maria", "Finance", 3000),  \
    ("James", "Sales", 3000),    \
    ("Scott", "Finance", 3300),  \
    ("Jen", "Finance", 3900),    \
    ("Jeff", "Marketing", 3000), \
    ("Kumar", "Marketing", 2000),\
    ("Saif", "Sales", 4100) \
  )
 
columns= ["employee_name", "department", "salary"]

df_window = spark.createDataFrame(data = simpleData, schema = columns)
```

```python
# row_number based on salary
windowSpec  = Window.orderBy(col("salary").desc())
df_window.withColumn("row_number", row_number().over(windowSpec)).show()
```

```python
# row_number based on window orderby salary decreasing
w = Window.partitionBy("department").orderBy("salary")
df_window.withColumn("row_number", row_number().over(w)).show()
```

```python
# row_number based on window orderby salary increasing
# rank gives same value to "Sales" partition based on tie in the salary values = 3000
w = Window.partitionBy("department").orderBy(col("salary").desc())
df_window.withColumn("row_number", row_number().over(w))\
         .withColumn("rank", rank().over(w))\
         .withColumn("dense_rank", dense_rank().over(w)).show()
```
