```python
from pyspark.sql.functions import lit
```

```python
data_1 = [(1,"Rohan"), (2,"Rohit")]
schema_1 = ["id","name"]
df_1 = spark.createDataFrame(data = data_1, schema = schema_1)
df_1.show()
```

```python
data_2 = [(1,"sortMerge","join"),(2,"broadcast","join"),(3,"skew","join"),(4,"range","join"),(5,"broadcastNestedLoop","join")]
schema_2 = ["join_id", "join_type","operation name"]
df_2 = spark.createDataFrame(data = data_2, schema = schema_2)
df_2.show()
```

```python
df_1 = df_1.withColumn("null column", lit("null"))
```

```python
df = df_1.union(df_2)
df.show()
```
