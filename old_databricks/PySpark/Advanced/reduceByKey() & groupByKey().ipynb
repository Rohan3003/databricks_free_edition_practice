```python
rdd1 = sc.parallelize([
    ("apples",1), ("oranges", 3), ("apples",10), ("mangoes", 12), ("oranges",13)
    ])
rdd_reduced = rdd1.reduceByKey(lambda x,y:x+y) # here the reduce function is lambda which add the values
rdd_reduced.collect()
```

```python
rdd2 = sc.parallelize([
    ("apples",1), ("oranges", 3), ("apples",10), ("mangoes", 12), ("oranges",13)
])
rdd_grouped = rdd2.groupByKey()
rdd_grouped.collect()
```

```python
# filtering data with low sales
sales_data = [
    ('product1', 100),
    ('product2', 200),
    ('product1', 300),
    ('product3', 400),
    ('product2', 500),
]

# Create an RDD from the sales data
sales_rdd = sc.parallelize(sales_data)
sales_rdd.collect()

sales_unique = sales_rdd.reduceByKey(lambda x,y : x+y)
sales_filtered = sales_unique.reduceByKey(lambda x,y: y >= 500)
sales_filtered.collect()
```
