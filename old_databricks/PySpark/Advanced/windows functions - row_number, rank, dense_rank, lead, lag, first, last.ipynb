Windows function - These are type of PySpark Transformation which creates window of records by grouping based on a key and apply certain logic(row_number, rank, dense_rank) with each window.

1. row_number()
2. rank()
3. dense_rank()
4. first() - returns first value of a column from each window.
5. last() - returns last value of a column from each window.
6. cum_dist()


```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank, dense_rank, col, first, last, desc
```

```python
data = [('Alice', 'Sales', 5000),
       ('Bob','Finance',6000),
       ('Charlie','Sales',5500),
       ('Luisa','HR',7000),
       ('Hector','Finance',2000),
       ("James","Sales",3000),
       ("Michael","Sales",4600),
       ("Robert","Sales",4100),
       ("Maria","HR",3000),
       ("Raman","Finance",3000),
       ("Scott","HR",3300),
       ("Jen","Finance",3900),
       ("Jeff","Marketing",3000),
       ("Kumar","Marketing",2000)]

columns = ['Name', 'Department', 'Salary']

df = spark.createDataFrame(data, columns)
df.printSchema()
df.show()
```

```python
df.orderBy('Department','Salary').show()
```

#### row_number()

```python
w = Window.partitionBy('department').orderBy('salary')
df.withColumn('row_number', row_number().over(w)).show()
```

```python
# to filter 2nd lowest salary from each department
df.withColumn('rn', row_number().over(w)).filter(col('rn') == 2)\
    .select('Name','Department','Salary')\
    .show()
```

### first() and last()

```python
df.orderBy('Department').show()
```

```python
w = Window.partitionBy('Department')
df.withColumn('first_value',first('Salary').over(w)).show()
```

```python
w = Window.partitionBy('Department')
df.withColumn('last_value',last('Salary').over(w)).show()
```

```python
# to get only first_value and last_value 
df.show()
WindowSpec = Window.partitionBy('Department')

df.withColumn('first_value',first('Salary').over(WindowSpec))\
    .withColumn('last_value',last('Salary').over(WindowSpec))\
        .drop('Name','Salary')\
            .distinct()\
                .show()
```
