```python
dbutils.widgets.text("mounting_point","")
dbutils.widgets.text("container","")
dbutils.widgets.text("storage_account","")
```

```python
mounting_point= dbutils.widgets.get("mounting_point")
container= dbutils.widgets.get("container")
storage_account = dbutils.widgets.get("storage_account")
print(mounting_point)
```

```python
#create temporary mount
configs = {"fs.azure.account.auth.type": "OAuth",
           "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
           "fs.azure.account.oauth2.client.id": dbutils.secrets.get(scope="databricks-secret-scope",key="adlsappid"),
           "fs.azure.account.oauth2.client.secret": dbutils.secrets.get(scope="databricks-secret-scope",key="adlssecret"),
           "fs.azure.account.oauth2.client.endpoint": f'https://login.microsoftonline.com/{dbutils.secrets.get(scope="databricks-secret-scope",key="adlstenant")}/oauth2/token'}
print(mounting_point)
#mount file system to get yaml config file
dbutils.fs.mount(
  source = f'abfss://{container}@{storage_account}.dfs.core.windows.net/',
  mount_point = mounting_point,
  extra_configs = configs)
```
