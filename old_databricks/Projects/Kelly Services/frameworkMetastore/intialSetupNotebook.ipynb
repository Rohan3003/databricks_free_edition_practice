#### Creating Metastore Database

```python
%sql
CREATE SCHEMA IF NOT EXISTS aceCurated
COMMENT 'Curation layer database for ACE data'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS aceRaw
COMMENT 'Raw layer database for ACE data'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS aceMetastore
COMMENT 'Metastore database for ACE data'
```

#### 1. jobControl Delta table

```python
spark.sql(f"""CREATE TABLE IF NOT EXISTS aceMetastore.jobControl
(sourceSystem STRING,
lastRunDate TIMESTAMP, 
currentRunDate TIMESTAMP) USING DELTA LOCATION '{jobControlPath}'""")
```

#### 2. pipelineAudit Delta table

```python
spark.sql(f"""CREATE TABLE IF NOT EXISTS aceMetastore.pipelineAudit(
 sourceSystem STRING ,
 sourceObjectName STRING,
 stageName STRING,
 stageOrder INTEGER,
 loadType STRING,
 adbPipelineID STRING,
 stageObjectName STRING,
 stageStartTs TIMESTAMP,
 stageEndTs TIMESTAMP,
 srcStartTs TIMESTAMP,
 srcEndTs TIMESTAMP,
 stageStatus STRING,
 errorLog STRING
 
 )
USING DELTA
LOCATION '{pipelineAuditPath}'""")
```

#### 3. dataQualityResult Delta table

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS aceMetastore.dataQualityResult(
  adbPipelineID STRING,
  ruleID INT,
  mappingID INT,
  ruleName STRING,
  expectedResult STRING,
  actualResult STRING,
  summary STRING,
  columnName STRING,
  dqStartTs TIMESTAMP,
  dqEndTs TIMESTAMP,
  dqStatus STRING,
  alertStatus STRING
)
USING DELTA
LOCATION '{dqResultPath}'""")
```

#### 4. adhocRunTable Delta table

```python
spark.sql(f"""CREATE TABLE IF NOT EXISTS aceMetastore.adhocRunTable
(sourceSystem STRING,
sourceObjectName STRING,
adhocRunFlag STRING)
USING DELTA
LOCATION '{adhocRunPath}'""")
```
