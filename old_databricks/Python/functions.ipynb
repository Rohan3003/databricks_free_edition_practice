### kwargs and args

1. *args is used to pass a variable number of non keyworded argument to a function

2. **kwargs is used to pass a variable number of keyword arguments to a function without specifying keyword argument

```python
def my_function_1(*args):
    for arg in args:
        print(arg)
        
my_function_1(1,2,3) # argument is passed to the function as a tuple
```

```python
def my_function_2(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} = {value}")
        
my_function_2(name="John", age=30, city="new York")  # arguments passed as a dictionary 
```
