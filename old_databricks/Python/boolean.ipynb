### boolean datatype

The boolean value of all empty iterable -list, set, tuples, string is False. <br>
The boolean value of all interger, float & complex number with 0, 0.0, 0j is also False. <br>

So, it is intutive to say that all non-zero values and non empty iterables will have truth value as True.

```python
# boolean value for an empty list
print(bool([]))

# boolean value of an empty string
print(bool(""))
      
# boolean value of None
print(bool(None))

# boolean value of int, float and complex
print(bool(0))
print(bool(0.0))
print(bool(0j))
```

```python
# boolean value for an non-empty list
print(bool([1,2]))

# boolean value of an non-empty string
print(bool("String"))

# boolean value of int, float and complex
print(bool(1))
print(bool(1.0))
print(bool(1j))
```

### any() and all() functions

any(iterable)

all(iterable)

```python
# Boolean checks 

money = True
wife = True
house = True
food = True

regs = [money, wife, house, food]

if all(regs):
    print('Good Job, All booleans are true')
else:
    print('Something is wrong, All booleans are not true')
```

```python
# Boolean checks 

money = False
wife = True
house = True
food = True

regs = [money, wife, house, food]

if all(regs):
    print('Good Job, All booleans are true')
else:
    print('Something is wrong, All booleans are not true')
```

```python
# Boolean checks 
money = False
wife = True
house = True
food = True

regs = [money, wife, house, food]

if any(regs):
    print('Good Job, All booleans are true')
else:
    print('Something is wrong, All booleans are not true')
```
