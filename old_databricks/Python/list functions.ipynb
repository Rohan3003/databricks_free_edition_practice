#### all() , any() for list

```python
list1 = ['python' ,  'javascript', 'csharp', 'go', 'c', 'c++']
list2 = ['csharp1' , 'go', 'python']

#all() is used to find if elements of list1 is in list2 or not
check = all(items in list1 for items in list2)
check
```

```python
list1 = ['python' ,  'javascript', 'csharp', 'go', 'c', 'c++']
list2 = ['csharp' , 'go', 'python', 'javascript']

#all() is used to find if elements of list1 is in list2 or not
check = all(items in list1 for items in list2)
check
```
