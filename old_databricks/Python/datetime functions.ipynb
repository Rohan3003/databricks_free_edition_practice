### strftime()

https://www.programiz.com/python-programming/datetime/strftime

```python
from datetime import datetime
```

```python
now = datetime.now()
print(now)
type(now)
```

```python
year = now.strftime('%Y')
year
```

```python
month = now.strftime('%m')
month
```

```python
day = now.strftime('%d')
day
```

```python
time = now.strftime('%H:%M:%S')
time
```

```python
# to get the timestamp to string 
timestamp_string = now.strftime('%y-%m-%d %H:%M:%S')
timestamp_string
```

```python
# to get the date from a timestamp to string 
date = now.strftime('%y-%m-%d')
date
```

```python
full_month = now.strftime('%B')
full_month
```
