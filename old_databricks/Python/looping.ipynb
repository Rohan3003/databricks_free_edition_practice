### Loops
Looping helps to iterate over a list, tuple, string, dictionary, dataframe etc
1. break - it breaks the loop when the condition is satisfied
2. continue - it will skip that particular iteration and run other iterations
3. pass - pass will simply do nothing 

### break

```python
for num in range(0,10):
    if num == 5:
        break
    print(f'Iteration : {num}')
```

```python
for k in range(0,3):
    print(f'This is outer loop with iteration {k}')
    for x in range(0,5):
        if x == 2:
            break
        print(f'This is inner loop with iteration {x}')
    print('----------------------------------------')
```

#### real world example : phone unlock
A real world example for break will be unlocking the phone with code.

```python
for chance in range(1,4):
    password = input('Enter password')
    if password == 'MyLock':
        print('Phone unlocked')
        break
    print(f'wrong password attempt: {chance}')
```

```python
for num in range(0,10):
    if num == 5:
        continue
    print(f'Iteration : {num}')
```

```python
for num in range(0,10):
    if num == 5:
        pass
    print(f'Iteration : {num}')
```
