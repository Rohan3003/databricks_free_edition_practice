```python
# Configure SFTP connection details
host = "sftp_host"
port = 22
username = "your_username"
password = "your_password"
sftp_path = "/path/to/remote/file.csv"
path = f"sftp://{username}:{password}@{host}:{port}:{sftp_path}"

# Load data from SFTP into a PySpark DataFrame
df = spark.read.format("csv") \
    .option("header", "true") \
    .load(path)

# Write the DataFrame to Delta Lake
delta_path = "/delta/path"
df.write.format("delta").save(delta_path)
```
