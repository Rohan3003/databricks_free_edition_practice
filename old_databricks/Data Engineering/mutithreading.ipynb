To run mutiple notebooks in parallel using multithreading

```python
from concurrent.futures import ThreadPoolExecutor

path = [
    '/Data Engineering/sample notebook'
]
def run_notebook(path):
    try:
        dbutils.notebook.run(path=path, timeout_seconds=120)
        print(f'Notebook {path} executed successfully')
    except Exception as e:
        print(f'Error executing notebook {path} : {e}')

try:
    with ThreadPoolExecutor(max_workers=3) as executor:
        executor.map(run_notebook, path)
except Exception as e:
    print(e)
```
