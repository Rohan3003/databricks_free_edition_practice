```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "Alice", "IT", 60000, 25),
    (2, "Bob", "HR", 50000, 30),
    (3, "Charlie", "IT", 70000, 28),
    (4, "David", "Finance", 55000, 35),
    (5, "Eve", "HR", 65000, 29),
    (6, "Frank", "Finance", 75000, 40),
    (7, "Grace", "IT", 80000, 32)
]

columns = ["emp_id", "name", "department", "salary", "age"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
df.groupBy("department").agg({"salary":"avg", "age" : "avg"}).show()
```
