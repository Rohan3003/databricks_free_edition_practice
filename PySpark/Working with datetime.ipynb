# Master dataFrame

```python
from pyspark.sql import functions as F

data = [
    # id, joining_date, last_login_ts, event_date_str, timezone, status
    
    (1, "2020-02-29", "2024-03-01 10:15:30", "2024/03/01", "UTC", "active"),  # leap year
    (2, "2019-12-31", "2024-01-01 00:00:00", "01-01-2024", "IST", "inactive"), # year boundary
    (3, "2021-01-15", "2024-03-31 23:59:59", "2024-03-31", "UTC", "active"),   # month end
    (4, "2022-06-10", "2024-04-01 00:00:01", "2024-04-01", "PST", "active"),   # month start
    (5, None, "2024-03-15 12:00:00", "15/03/2024", "IST", "inactive"),         # null joining
    (6, "2023-02-28", None, "2024-02-29", "UTC", "active"),                    # null timestamp + leap
    (7, "2023-03-01", "2024-03-01 08:30:00", "invalid_date", "UTC", "active"), # invalid string
    (8, "2020-01-01", "2024-12-31 23:59:59", "2024-12-31", "IST", "inactive"), # year end
    (9, "2024-03-01", "2024-03-01 00:00:00", "2024-03-01", "UTC", "active"),   # same day
    (10, "2018-07-20", "2024-03-15 14:45:00", "2024-03-15", "CET", "active")   # long tenure
]

columns = ["id", "joining_date", "last_login_ts", "event_date_str", "timezone", "status"]

df_master = spark.createDataFrame(data, columns)

# Convert to proper types
#df = df.withColumn("joining_date", F.to_date("joining_date")) \
 #       .withColumn("last_login_ts", F.to_timestamp("last_login_ts"))

df_master.show(truncate=False)
```

# to_date(), to_timestamp()

```python
df_1 = df_master.withColumn("joining_date", F.to_date("joining_date"))\
                .withColumn("last_login_ts", F.to_timestamp("last_login_ts"))
df_1.show()
```
