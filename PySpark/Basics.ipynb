## when() & otherwise()

```python
df_master.show(5)
```

```python
df_conditional = df_master.select("name", "age", F.when(df_master.age >=35, "Senior")\
                                                  .when(df_master.age >=50, "Super Senior")
                                                  .otherwise("Junior")\
                                                  .alias("age_group"))
df_conditional.show(5)
```

```python
df_conditional_1 = df_master.withColumn("random_value", F.when(F.col("salary") > 50000, F.rand()))
df_conditional_1.show(10)
```

# Working with Arrays

## flatten()

```python
data = [
    (1, [[1, 2], [3, 4]]),
    (2, [[5, 6], [7, 8]])
]

df = spark.createDataFrame(data, ["id", "nums"])
df.show(truncate=False)
```

```python
df.select("id", flatten("nums").alias("flatten_applied")).show()
```

```python
df.withColumn("flatten_applied", flatten("nums")).show()
```

```python
from pyspark.sql.functions import explode
df_exploded = df_master.withColumn("skillset", explode("skills"))\
         .drop("skills")
df_exploded.show()
```
